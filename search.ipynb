```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv("crx.csv")
df.sample(5)
```

```python
df.drop('A14', axis=1, inplace=True)
df.drop('A11', axis=1, inplace=True)

df.sample(5)
```

```python
df1 = df[(df.A1 != '?') & (df.A2 != '?') & (df.A4 != '?') & (df.A5 != '?') & (df.A6 != '?') & (df.A7 != '?')]
df1.A2 = pd.to_numeric(df1.A2)
```

```python
df1['A1'].replace({'a': 1,'b': 0},inplace=True)
df1['A9'].replace({'t': 1,'f': 0},inplace=True)
df1['A10'].replace({'t': 1,'f': 0},inplace=True)
df1['A12'].replace({'t': 1,'f': 0},inplace=True)
df1['T'].replace({'+': 1,'-': 0},inplace=True)
```

```python
df1.sample(6)
```

```python
df2 = pd.get_dummies(data=df1, columns=['A4','A5', 'A6', 'A7', 'A13'])
```

```python
df2.sample(5)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df2[df2.columns] = scaler.fit_transform(df2[df2.columns])
print("done")
```

```python
df2.sample(5)
```

```python
X = df2.drop('T',axis='columns')
y = df2['T']
```

```python
X.shape[1]
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import regularizers

def train(X_train, y_train, activation_f, hidden, loss, opt):

    layers = [];

    layers.append(keras.layers.Dense(
            X_train.shape[1], 
            input_shape=(X_train.shape[1],),
            activation=activation_f,
            activity_regularizer=regularizers.l2(hidden[0][1])
        ))
    for i in hidden:
        layers.append(keras.layers.Dense(
            i[0],
            activation=activation_f,
            activity_regularizer=regularizers.l2(i[1])
        ))
    
    layers.append(keras.layers.Dense(
            1, 
            activation='sigmoid'
        ))
    model = keras.Sequential(layers)

    # opt = keras.optimizers.Adam(learning_rate=0.001) binary_crossentropy

    model.compile(optimizer=opt,
                  loss=loss,
                  metrics=['accuracy', 'Precision', 'Recall'])

    model.fit(X_train, y_train, epochs=50,verbose=0)
    
    return model
```

```python
def evaluate(model, X_test, y_test):
    model.evaluate(X_test, y_test)
```

```python
def prediction(model, X_test):
    yp = model.predict(X_test)
    y_pred = []
    for element in yp:
        if element > 0.5:
            y_pred.append(1)
        else:
            y_pred.append(0)
            
    return y_pred
```

```python
from sklearn.metrics import confusion_matrix , classification_report

def report(y_test, y_pred):
    print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import f1_score
def f1score(y_test, y_pred):
    f1 = f1_score(y_test, y_pred)
    # print(f1)
    return f1
```

```python
def modelF1(X_train, y_train, X_test, y_test, activation_f, hidden, loss, opt):
    # print(' ------------ Training --------------')
    model = train(X_train, y_train, activation_f, hidden, loss, opt)
    # print(' ------------ Evaluation --------------')
    # evaluate(model, X_test, y_test)
    # print(' ------------ Prediction --------------')
    y_pred = prediction(model, X_test)
    # print(' ------------ Report --------------')
#     report(y_test, y_pred)
    # print(' ------------ F1-Score --------------')
    return f1score(y_test, y_pred)
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=5)

activation_f = ['relu']
hidden =[[[3,0]], [[4,0]], [[5,0]], [[6,0]], [[7,0]], [[8,0]], [[13,0], [5,0]], [[12,0], [4,0]], [[11,0], [4,0]], [[11,0], [3,0]], [[10,0], [3,0]]]

loss = ['binary_crossentropy', 'mean_squared_error']

lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-1,
    decay_steps=10000,
    decay_rate=0.9)

opt = [keras.optimizers.Adam(learning_rate=lr_schedule), keras.optimizers.SGD(learning_rate=lr_schedule)]
maxf1 = 0
maxf1_config = ''

for i in range(44):
    f1_total = 0
    print(f'Iteration - {i + 1}/44')
    for train_index, test_index in folds.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        f1_total = f1_total + modelF1(X_train, y_train, X_test, y_test, activation_f[0], hidden[i % 11], loss[(i//11) % 2], opt[(i//22) % 2])

    f1_avg = f1_total / 5
    print(f'F1: {f1_avg} - Activation function: {activation_f[0]}, hidden layers: {len(hidden[i % 11])}, hidden: {hidden[i % 11]}, loss: {loss[(i//11) % 2]}, opt: {(i//22) % 2}')
    if(f1_avg > maxf1):
        print("---- MAX changed -----")
        maxf1 = f1_avg
        maxf1_config = f'Activation function: {activation_f[0]}, hidden layers: {len(hidden[i % 11])}, hidden: {hidden[i % 11]}, loss: {loss[(i//11) % 2]}, opt: {(i//22) % 2}'

print(f'--------- F1 - {maxf1}, Configuration - {maxf1_config}')
```
